```python
# plotting a bar chart for Gender and it's count
ax = sns.countplot(x = 'Gender' ,data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Popularity_Rating'].sum().sort_values(by='Popularity_Rating', ascending=False)
```

```python
# plotting a bar chart for Gender vs popularity_rate
choosen_method =df.groupby(['Gender'], as_index=False)['Popularity_Rating'].sum().sort_values(by='Popularity_Rating', ascending=False)
sns.barplot(x = 'Gender',y= 'Popularity_Rating' ,data = choosen_method)
```

```python
#From the above graphs we can see most of the buyers are purchasing unisex productes according to the popularity_rating
```

```python
ax = sns.countplot(data = df, x = 'Age_Group')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#From above graph we can see the highest rate of buyer are of ADULT age group
```

```python
sales_state = df.groupby(['Brand'], as_index=False)['SPF_Rating'].sum().sort_values(by='SPF_Rating', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'Brand',y= 'SPF_Rating')
```

```python
ax = sns.countplot(x = 'Brand' ,data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#From above graph we can see that SPF_Rating is higher to ponds product compare to other products and purchasing rate is higher to himalaya 
```

```python
 ax = sns.countplot(x = 'Category' ,data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Usage = df.groupby(['Category', 'Usage_Frequency'], as_index=False)['Average_Price'].sum().sort_values(by='Average_Price', ascending=False).head(10)

sns.set(rc={'figure.figsize':(16,8)})
sns.barplot(data = Usage, x = 'Category',y= 'Average_Price', hue= 'Usage_Frequency')
```

```python
#From the above graph we can see skincare in daily is the most costly compare to other catogary and according to usage_frequency people is more care to bodycare
```

```python
sales_state = df.groupby(['Brand'], as_index=False)['Average_Price'].sum().sort_values(by='Average_Price', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'Brand',y= 'Average_Price')
```

```python
#From the above graph we can understand himalaya brand is most earning company in the cosmetics field
```
